# Q1) Which countries have the highest and lowest average population growth rate in 1961?

# Analysis:

```python
# Find the country with the highest average population growth rate in 1961
highest_growth_country_1961 = df.loc[df['1961'].idxmax()]['country_name']

# Find the country with the lowest average population growth rate in 1961
lowest_growth_country_1961 = df.loc[df['1961'].idxmin()]['country_name']

print("Country with the highest average population growth rate in 1961:", highest_growth_country_1961)
print("Country with the lowest average population growth rate in 1961:", lowest_growth_country_1961)
```

# Visualization:

```python
# Plot the top 10 countries with highest average population growth rate in 1961
plt.figure(figsize=(20, 10))  # Set the figure size  20 = width  , 10 = height

plt.title('Top 10 Countries with Highest Average Population Growth Rate in 1961')  # Set the title of the plot

plt.xlabel('Country')  # Set the label for the x-axis
plt.ylabel('1961')  # Set the label for the y-axis

# Bar chart showing average population growth rate for each country in 1961
sns.barplot(x=df['country_name'].head(10), y=df['1961'].head(10))  # Plot the bar chart

plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()  # Show the plot
```

# Q5) What is the correlation between the population growth rates of China, India, and the United States for the years 2019, 2020, and 2021?


# Analysis:

```python
# Select three countries
country1 = 'China'
country2 = 'India'
country3 = 'United States'

# Select three years
years = ['2019', '2020', '2021']

# Filter the dataframe to include only the selected countries and years
selected_countries_years_df = df[df['country_name'].isin([country1, country2, country3])]
selected_countries_years_df = selected_countries_years_df[['country_name'] + years]

# Calculate the correlation matrix
correlation_matrix = selected_countries_years_df.iloc[:, 1:].corr()
```

# Visualization:

```python

# Plot the correlation matrix as a heatmap
plt.figure(figsize=(20, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Population Growth Rate for Selected Countries and Years')
plt.xlabel('Year')
plt.ylabel('Year')
plt.show()
```

# Q6) How does the population growth rate vary between high-income and low-income countries over the years?

# Analysis:

```python
# Filter the dataframe to include only high-income and low-income countries
high_income_countries = ['United States', 'Germany', 'Japan']
low_income_countries = ['India', 'Nigeria', 'Bangladesh']

high_income_df = df[df['country_name'].isin(high_income_countries)]
low_income_df = df[df['country_name'].isin(low_income_countries)]

# Calculate the median population growth rate for high-income and low-income countries by year
high_income_median_growth = high_income_df.iloc[:, 2:].median(axis=0)
low_income_median_growth = low_income_df.iloc[:, 2:].median(axis=0)
```

# Visualization:

```python

# Plot the median population growth rate for high-income and low-income countries over the years
plt.figure(figsize=(20, 8))
plt.plot(high_income_median_growth.index, high_income_median_growth.values, marker='o', color='blue', linestyle='-', label='High Income')
plt.plot(low_income_median_growth.index, low_income_median_growth.values, marker='o', color='red', linestyle='-', label='Low Income')
plt.title('Median Population Growth Rate: High-Income vs Low-Income Countries')
plt.xlabel('Year')
plt.ylabel('Median Population Growth Rate')
plt.xticks(rotation=90)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

# Q7) How does the population growth rate in Europe compare to that in North America over the years?

# Analysis:

# Q12) Which countries have population growth rates that are significantly different from the global average in the most recent year?


# Analysis:

```python
most_recent_year = df.columns[-1]
df['z_score'] = (df[most_recent_year] - df[most_recent_year].mean()) / df[most_recent_year].std()

outliers = df[(df['z_score'] > 2) | (df['z_score'] < -2)]

print("Countries with significant population growth rate differences in", most_recent_year, ":")
print(outliers[['country_name', most_recent_year, 'z_score']])
```

# Visualization:

```python

# Visualization: Plot z-scores
plt.figure(figsize=(10, 6))
plt.hist(df['z_score'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Distribution of Z-Scores for Population Growth Rate')
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.show()
```

# Q13) What are the countries with significant population growth rate differences, and why might these differences occur?

# Q15) Which of the five selected countries have population growth rates that are unusually high or low compared to the majority of countries?

# Analysis:

```python
# Select the population growth rates of five specific countries
selected_countries = ['China', 'India', 'United States', 'Indonesia', 'Pakistan']
selected_growth_rates = df[df['country_name'].isin(selected_countries)].set_index('country_name').iloc[:, 2:]

# Calculate the IQR for each country
Q1 = selected_growth_rates.quantile(0.25)
Q3 = selected_growth_rates.quantile(0.75)
IQR = Q3 - Q1

# Define outliers for each country
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = (selected_growth_rates < lower_bound) | (selected_growth_rates > upper_bound)

# Create a DataFrame containing only the outliers
outlier_values = selected_growth_rates[outliers]
```
